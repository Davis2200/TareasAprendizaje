```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


plt.style.use('ggplot')
sns.set_palette("husl")
%matplotlib inline

tickets = pd.read_csv('tickets_productos.csv')
```

```python
print("Dimensiones del dataset:", tickets.shape)
print("\nPrimeras filas:")
tickets.head()
```

```python
ticket_numero_df = tickets[['ticket_id', 'producto']].copy()
ticket_numero_df['primer_numero'] = tickets['producto'].str.extract(r'(\d+)')[0]

#tickets[tickets['producto'].str.contains(r'^\d', na=False, regex=True)]
#cantidadproducto = tickets['producto'].str.extract(r'^(\d+)')
```

```python
plt.figure(figsize=(14, 8))
plt.scatter(tickets['producto'], tickets['cantidad'], alpha=0.6, color='steelblue')
plt.title('Diagrama de Dispersión: Productos vs Cantidad')
plt.xlabel('Producto')
plt.ylabel('Cantidad')
plt.xticks(rotation=90)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
