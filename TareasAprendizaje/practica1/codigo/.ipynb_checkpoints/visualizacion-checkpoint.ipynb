```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import re


plt.style.use('ggplot')
sns.set_palette("husl")
%matplotlib inline

tickets = pd.read_csv('tickets_productos.csv')
```

```python
print("Dimensiones del dataset:", tickets.shape)
print("\nPrimeras filas:")
tickets.head()
```

```python
#Extraccion de producto con id del ticket
ticket_numero_df = tickets[['id_ticket', 'producto']].copy()
#Extraer los productos unicos con su frecuencia
productos_unicos = ticket_numero_df['producto'].value_counts().index.tolist()
frecuencias = ticket_numero_df['producto'].value_counts().values.tolist()
#Creación del dataframe de productos con su frecuencia
frecuencia_producto = pd.DataFrame({'producto': productos_unicos, 'frecuencia': frecuencias})

```

```python
#en este dataframe tenemos la cantidad de cada producto por ticket
cantidad_producto = tickets[['id_ticket', 'producto', 'cantidad', 'precio']].copy()
  
```

```python
plt.figure(figsize=(14, 8))
plt.scatter(tickets['producto'], tickets['cantidad'], alpha=0.6, color='steelblue')
plt.title('Diagrama de Dispersión: Productos vs Cantidad')
plt.xlabel('Producto')
plt.ylabel('Cantidad')
plt.xticks(rotation=90)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
#aqui podemos ver el total de piezas vendidas por cada producto
piezas_por_producto = tickets.groupby('producto')['cantidad'].sum().reset_index()
```

```python
plt.figure(figsize=(16, 8))
plt.bar(piezas_por_producto['producto'], piezas_por_producto['cantidad'], color='steelblue', alpha=0.7)
plt.title('Piezas vendidas por producto')
plt.xlabel('Producto')
plt.ylabel('Cantidad de piezas vendidas')
plt.xticks(rotation=90)
plt.grid(True, axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```
